```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_dki_2014.csv')
```

```python
df = data.copy()
```

```python
df
```

```python
df.head()
```

```python
mean_1 = df['laki-laki_wna'].mean()

mean_2 = df['perempuan_wna'].mean()
```

```python
df['laki-laki_wna'] = df['laki-laki_wna'].fillna(value=np.round(mean_1))
```

```python
df['perempuan_wna'] = df['perempuan_wna'].fillna(value=np.round(mean_2))
```

```python
df
```

```python
df['total_wni'] = df['laki-laki_wni'] + df['perempuan_wni']

```

```python
df['total_wna'] = df['laki-laki_wna'] + df['perempuan_wna']

df['total_populasi'] = df['total_wni'] + df['total_wna']

df
```
